```python
pilot_users['Gender'] = ['F','M','M','F','M','M','M','M','F','M']
pilot_users
```

```python
import numpy as np
def stratified_assignment(group):
    n = len(group)
    group = group.sample(frac=1).reset_index(drop=True)  # Shuffle within gender
    group['Group'] = np.tile([1, 0], n // 2 + 1)[:n]  # Assign alternating 0s and 1s
    return group

students = students.groupby('Gender', group_keys=False).apply(stratified_assignment)
```

```python
students.groupby('Gender').Group.value_counts()
```
